## cut()

````python
bins = [0, 100, 200, 300, df_copy['운동량'].max()] # 내가 보고싶은 구간(범위)를 리스트로 지정
````

````python
pd.cut(df_copy['운동량'], bins) # 내가 나눌 구간을 리스트로
# 2, 4번과 같은 값이 이상치 - 추후 제거, 대체 등의 작업이 필요함
````

````python
pd.cut(df_copy['운동량'], bins=10) # 가장 작은 것부터 가장 큰 것을 10씩 나누어서 구분함
# 앞에서 n% 구간을 찾기
pd.cut(df_copy['운동량'], bins=10).value_counts()
````

````python
pd.qcut(df_copy['운동량'], q=10) # qcut은 비율(백분위)대로 나누기
pd.qcut(df_copy['운동량'], q=10).value_counts()
````

````python
qcut_bins = [0, 0.2, 0.8, 1] # 0% - 20%, 20% - 80%, 80%- 100% 구간 나눔
pd.qcut(df_copy['운동량'], qcut_bins).value_counts()
# 맨 위 값이 중간값
````

````python
qcut_bins = [0, 0.2, 0.8, 1] # 0% - 20%, 20% - 80%, 80%- 100% 구간 나눔
# labels는 글자
df_copy['운동량_한글'] = pd.qcut(df_copy['운동량'], qcut_bins, labels=['적음', '보통', '많음'])
df_copy.head(2)
````
