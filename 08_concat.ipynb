## concat()

```python
pd.concat([df1, df2])
# 리스트에 2개 이상의 데이터프레임 넣을 수 있어용
```

## 열병합

```python
sales1 = df1.iloc[:, :5]
# 위에서 아래까지 전체
# 앞에서부터 5번째까지 column

sales2 = df2.iloc[:, 5:]
# 범위로 묶을 때는 iloc, 대부분의 상황에서는 loc(위치 직접 지정)를 더 많이 사용
```

## merge()

```python
info = {
    '고객명': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']
}

money = {
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']
}
```

```python
# 딕셔너리를 데이터프레임으로 만들기
df1 = pd.DataFrame(info)
df2 = pd.DataFrame(money)
```

```python
pd.concat([df1, df2])
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2) # 교집합(innerjoin) 데이터만 만들어줌
```

```python
pd.merge(df1, df2, how='left')
# how값에 병합할 기준(교집합, 합집합, 왼쪽, 오른쪽)을 넣어줌
```

```python
pd.merge(df1, df2, how='right')
```

```python
pd.merge(df1, df2, how='outer')
```

```python
# 같은 value를 가진 column의 이름이 다를 때
info = {
    '이름': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']
}

money = {
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']
}
```

```python
# 딕셔너리를 데이터프레임으로 만들기
df1 = pd.DataFrame(info)
df2 = pd.DataFrame(money)
```

```python
pd.merge(df1, df2, left_on='이름', right_on='고객명')
# 내가 직접적으로 지정해줌
# 왼쪽에서 기준이 되는 컬럼, 오른쪽에서 기준이 되는 컬럼
```
