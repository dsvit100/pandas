### 결측 데이터 필터링

````python
# 필터링 한 후에 값 변경

cond = df_copy['age'].isnull()
# df_copy.loc[cond]  # null값이 있는 값들을 모두 가져옴

df_copy.loc[cond, 'age'] = 30  # age라는 컬럼에 30살을 넣을거양
df_copy.tail()
````

## fillna()
- 결측값을 특정값으로 채우는 함수
- 계산 결과를 보여주는 함수이기 때문에 값을 재할당 하려면 덮어쓰기 해줘야 함

````python
df_copy = df.copy()
````

````python
df_copy.tail()
````

````python
df_copy['age'].fillna(100)
# 파일명[칼럼명] 필드 채워주기
````

````python
df_copy.tail
````

````python
df_copy.info()
````

````python
df_copy['deck'].value_counts()
````

````python
df_copy['deck'].fillna('C')
````

````python
df_copy['deck'].fillna('Z') # 오류남
# 내가 지정하지 않은 카테고리에 넣을 수가 없음
# 먼저 카테고리를 추가해 주어야함

## 카테고리란?
# 현재 deck은 ['A'...'G']까지의 카테고리가 있음
# 이는 사람이 보기 편하게 글자로 표시되지만
# 판다스 내부적으로는 숫자로 인식하여 처리되기 때문에 속도가 빠름
# (예) 혈액형
````

````python
df_copy['deck'] = df_copy['deck'].cat.add_categories('Z')
df_copy['deck'].fillna('Z')
````

````python
# age 결측치를 평균값으로 채우기
age_mean = df_copy['age'].mean()
````

````python
df_copy['age'] = df_copy['age'].fillna(age_mean)
# 'age' 재할당
````

````python
df_copy.info()
````

````python
df_copy = df.copy()
````

````python
# age 결측치를 중앙값으로 채우기
df_copy['age'] = df_copy['age'].fillna(df_copy['age'].median())
df_copy.tail()
````
